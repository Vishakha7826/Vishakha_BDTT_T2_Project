```

#Question 2 - Listing all the types along with frequencies with most frequent order
rdd_clinicaltrial_2021 = sc.textFile("FileStore/tables/clinicaltrial_2021.csv")
rdd_clinicaltrial_2021_split = rdd_clinicaltrial_2021.map(lambda x:  x.split("|"))
rdd_clinicaltrial_2021_type = rdd_clinicaltrial_2021_split.map(lambda x: x[5])
rdd_clinicaltrial_2021_join = rdd_clinicaltrial_2021_type.map(lambda word: (word, 1))
rdd_clinicaltrial_2021_count = rdd_clinicaltrial_2021_join.reduceByKey(lambda x,y : x+y)
rdd_clinicaltrial_2021_show = rdd_clinicaltrial_2021_count.sortBy(lambda x : [-x[1], x[0]])
rdd_clinicaltrial_2021_show.take(4)
```

```
#Question 3 - Top five conditions with their frequencies

rdd_clinicaltrial_2021 = sc.textFile("FileStore/tables/clinicaltrial_2021.csv")
rdd_clinicaltrial_2021_data = rdd_clinicaltrial_2021.map(lambda x: x.split('|'))
rdd_clinicaltrial_2021_term = rdd_clinicaltrial_2021_data.map(lambda x: x[7])
 
clinicaltrial_1 = rdd_clinicaltrial_2021_term.first()
 
rdd_clinicaltrial_2021_filter= rdd_clinicaltrial_2021_term.filter(lambda x: x != clinicaltrial_1)
rdd_clinicaltrial_2021_split = rdd_clinicaltrial_2021_filter.flatMap(lambda z: z.split(','))
rdd_clinicaltrial_2021_join = rdd_clinicaltrial_2021_split.map(lambda word: (word, 1))
rdd_clinicaltrial_2021_count = rdd_clinicaltrial_2021_join.reduceByKey(lambda x,y : x+y)
rdd_clinicaltrial_2021_sort = rdd_clinicaltrial_2021_count.sortBy(lambda x : [-x[1], x[0]])
 
clinicaltrial_2 = rdd_clinicaltrial_2021_sort.first()
 
clinicaltrial_2021_rdd = rdd_clinicaltrial_2021_sort.filter(lambda x: x != clinicaltrial_2)
clinicaltrial_2021_rdd.take(5)
```

```
#Question 5 -  Number of completed studies each month in a given year

RDD_data = rdd_clinicaltrial_2021.map(lambda x: x.split("|"))
rdd_clinicaltrial_2021_data = RDD_data.map(lambda x: (x[2], x[4]))
rdd_clinicaltrial_2021_filter = rdd_clinicaltrial_2021_data.filter(lambda x: 'Completed' in x)
rdd_clinicaltrial_2021_map = rdd_clinicaltrial_2021_filter.map(lambda x: x[1])
rdd_clinicaltrial_2021_filter2 = rdd_clinicaltrial_2021_map.filter(lambda x: '2021' in x)
rdd_clinicaltrial_2021_map2 = rdd_clinicaltrial_2021_filter2.map(lambda x: (x, 1))
rdd_clinicaltrial_2021_red = rdd_clinicaltrial_2021_map2.reduceByKey(lambda x,y: x+y)
```

```
from datetime import datetime
```

```
def months(date):
   date = datetime.strptime(date,'%b %Y')
   monthly_2021 = date.month
   return  monthly_2021
 
rdd_clinicaltrial_2021_map3 = rdd_clinicaltrial_2021_red.map(lambda line: (line[0],  months(line[0]),line[1]))
rdd_clinicaltrial_2021_sort = rdd_clinicaltrial_2021_map3.sortBy(lambda line: line[1])
rdd_clinicaltrial_2021_map4  = rdd_clinicaltrial_2021_sort.map(lambda line : (line[0].split(' ')[0],line[2]))
 
rdd_clinicaltrial_2021_map4 .take(12)
```
